## Variables and Types 
Variables store information. Python figures out the type automatically. 

```python
# Strings - Text that is always in quotes 
name = "Kamal"
city = "Guadalajara" 

# Integers - Whole numbers, no quotes

age = 35 
days_in_mexico = 23

# Floats - decimal numbers

rent_per_month = 750.50
exchange_rate = 21.35

# Booleans - True or False (capital T and F)

is_relocating = True
debt_cleared = False

print(type(name))
print(type(age))
print(type(rent_per_month))
print(type(is_relocating))
```

## Strings in Depth 

Strings are text and will be used extensively. They have many built-in functions.

```python
## String Operations

city = " Guadalajara " 

# Cleaning strings - common in real data

print(city.strip())       # Removes whitespace from both ends
print(city.lower())       # All lowercase 
print(city.upper())       # All uppercase 

# Checking content 

print("jara" in city)     # True - checks if "jara" is inside the string
print(city.startswith("Gua"))  # False - because of the spaces 
print(city.strip().startswith("Gua"))  # True - Strip first, then check

# Splitting strings 

sentence = "Guaadalajara, Chihuahua, Aguascalientes, Hermosillo, Mazatlan"
cities_list = sentence.split(", ") # Splits at every ", " 
print(cities_list)   # ['Guadalajara', 'Chihuahua', 'Aguascalientes', 'Hermosillo', 'Mazatlan'] 

# f-strings - the modern way to put variables inside text

name = "Kamal" 
destination = "Guadalajara" 
budget = 1500.75

print(f"Hi, I'm {name}. I'm planning to move to {destination} on a budget of £{budget} per month.")

# String indexing - every character has a position number (starting from 0)

word = "Python"
print(word[0])   # P
print(word[1])   # Y
print(word[-1])  # N (last character '-' denotes going backwards from end of the string)
print(word[0:3]) # Pyt - (characters 0, 1, 2 - NOT 3)
```

Most useful aspect of this for me. Cleaning strings - using print(city.strip()) to remove whitespaces. Checking content especially print(city.startwith()) or print(city.strip().startswith()) will be a useful search function within strings. 

F-strings will be crucial regarding efficiency in implementing variables within text and will enable greater legibility when reading through code.

## Lists - ordered collections of items

```python
# Creating a list

cities = ["Guadalajara", "Chihuahua", "Aguascalientes", "Hermosillo", "Mazatlan"] 

# Accessing items (counting starts at 0) 

print(cities[0])    # Guadalajara 
print(cities[2])    # Aguascalientes
print(cities[-1])   # Mazatlan (the last item) 

# How many items 

print(len(cities))  # 5 

# Adding items 

cities.append ("San Luis Potosi")
print(cities)   # Now has 6 cities 

# Removing items 

cities.remove ("Mazatlan") 
print(cities)  # Mazatlan is gone 

# Checking if something is in the list 

print("Hermosillo" in cities)  # True 
print("Monterrey" in cities)   # False 

# Sorting

cities.sort()
print(cities)  # Alphabetical order

# Slicing - getting a portion of the list 

print(cities[0:3])  # First 3 cities 
print(cities[2:])  # Everything from index 2 onwards

# Lists can hold different types (though usually they are kept the same) 

mixed = ["Kamal", 35, True, 1500.50]
print(mixed)
```

## Dictionaries

```python
## Dictionaries - Key value pairs 

# A dictionary about a city

guadalajara = { 
    "name": "Guadalajara", 
    "country": "Mexico",
    "state": "Jalisco",
    "avg_rent_usd": 650,
    "has_coworking": True,
    "timezone": "UTC-6"
}

# Accessing values by key

print(guadalajara["name"])  #  Guadalajara
print(guadalajara["avg_rent_usd"]) # 650

# Adding a new key value pair

guadalajara["population_millions"] = 5.2 
print(guadalajara)

# Changing a value

guadalajara["avg_rent_usd"] = 700
print(guadalajara["avg_rent_usd"])

# Getting all keys 

print(guadalajara.keys())

# Getting all values 

print(guadalajara.values())

# Checking if a key exists

print("name" in guadalajara)   # True
print("restaurants" in guadalajara) # False 

# Safe access with .get() - does not crash if key is missing 

print(guadalajara.get("restaurants", "no data"))  # Will return no data instead of crashing 
```

## Loops

```python
## Loops - doing something to every item

# Looping through a list 

cities = ["Guadalajara", "Chihuahua", "Aguascalientes", "Hermosillo", "Mazatlan"]

for city in cities: 
    print(f"I'd like to visit {city}")

# Variable name after for is my choice. City is appropriate in this case but it can be called anything. Such as:

for x in cities:   # this is exactly the same, its just less legible. 
    print(x) 

# Looping with a counter using enumerate()

for i, city in enumerate(cities):
    print(f"{i + 1}. {city}")

# Output 

# 1. Guadalajara
# 2. Chihuahua 
# 3. Aguascalientes
# 4. Hermosillo
# 5. Mazatlan

# Looping through a range of numbers

for number in range(5):
    print(number)

# Looping through a dictionary 

guadalajara = {"name": "Guadajara", "rent": 650, "timezone": "UTC-6"}

for key, value in guadalajara.items():
    print(f"{key}:{value}")
```

```python
## If / Else - Making decisions in code 

budget = 1500
rent = 650

# Simple if

if rent < budget:
    print("This city is affordable!")

# If / else 

if rent > budget:
    print("Too expensive")
else:
    print("Within budget")

# If / elif / else

remaining = budget - rent 

if remaining > 500:
    print(f"Comfortable, £{remaining} left after rent")
elif remaining > 200:
    print(f"Tight but manageable, £{remaining} left after rent")
elif remaining > 0:
    print(f"Very tight, only £{remaining} left after rent")
else:
    print("Cannot afford this city")

# Combining conditions with and / or

city = "Guadalajara"
has_coworking = True

if rent < budget and has_coworking:
    print(f"{city} is a food fit!")
    
if rent > 2000 or not has_coworking:
    print("Skip this city")

else:
    print("Worth considering")
```

## Combining everything - First programme

```python
## City Comparison Tool

# Data: A list of dictionaries (how real data often looks)
       
cities = [
    {
        "name": "Guadalajara",
        "rent_gbp": 658,
        "internet_reliable": True,
        "safety_rating": 4,
        "timezone": "UTC-6"
    },
    {
        "name": "Chihuahua",
        "rent_gbp": 667,
        "internet_reliable": True,
        "safety_rating": 5,
        "timezone": "UTC-6"
    },
    {
        "name": "Hermosillo",
        "rent_gbp": 466,
        "internet_reliable": False, 
        "safety_rating": 5,
        "timezone": "UTC-7"
    },
    {
        "name": "Aguascalientes",
        "rent_gbp": 389,
        "internet_reliable": True, 
        "safety_rating": 7,
        "timezone": "UTC-6"
    },
    {
        "name": "Mazatlan",
        "rent_gbp": 608,
        "internet_reliable": True,
        "safety_rating": 6,
        "timezone": "UTC-7"
    },
    {
        "name": "San Luis Potosi",
        "rent_gbp": 494,
        "internet_reliable": True,
        "safety_rating": 5,
        "timezone": "UTC-6"
    },
    {
        "name": "Monterrey",
        "rent_gbp": 745,
        "internet_reliable": True,
        "safety_rating": 5,
        "timezone": "UTC-6"
    }
]
        
# My Criteria 

max_budget = 700
need_internet = True 

# Loop through and evaluate each city 

print("=== City Evaluation ===\n")

affordable_cities = []   # Empty list that we will fill when going along

for city in cities:
    name = city["name"]
    rent = city["rent_gbp"]
    internet = city["internet_reliable"]
    safety = city["safety_rating"]

    print(f"--- {name} ---")
    print(f"Rent: £{rent}/month")

    if rent <= max_budget:
        print(f" Within budget (£{max_budget - rent} under)")
    else:
        print(f" Over budget by £{rent - max_budget}")
        continue          # No point checking more relevant to this city - so skip to next city

    # Check internet

    if need_internet and not internet:
        print(f" Internet not reliable, dealbreaker for remote work")
        continue
    else: 
        print(f" Internet: {'reliable' if internet else 'unreliable'}")

    # Check safety

    if safety >= 7:
        print(f" Safety: {safety}/10")
    else:
        print(f" Safety: {safety}/10 - worth researching more")

    # If we got here city passes all checks

    affordable_cities.append(name)
    print(f" {name} makes the shortlist!")

    print()   # Empty line between cities

# Summary

print("=== Results ===")
print(f" Cities that passed all criteria: {affordable_cities}")
print(f" Total options: {len(affordable_cities)}")
```

### What this programme does: 

This is a City Comparison Tool where a list of dictionaries is provided containing data for different cities in Mexico. This data includes name of the city, average rent, internet reliability, safety and timezone. Following this the variables are set (variables which can be changed). Focusing on two specifically - Max rent and requirement for reliable internet. Upon doing this - we create a loop to evaluate each city against these variables using if/ else statements. 

If a city fails a check such as rent being too high or internet being unreliable - the continue function skips the rest of the evaluation for that city and moves to the next city in the loop. 

Following this, any cities that pass these checks via loop evaluation have their names then appended to a separate results list which is printed at the end. With a total count of the cities that passed checks being completed through the len() function.
